```python
import uuid
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx


iteration = os.listdir("sampling/")[9]

folder = "sampling/"+iteration

df = pd.read_csv(folder+"/history.csv", index_col=0)
ground_truth_params = pd.read_csv(folder+"/true_params.csv",index_col=0)["0"].values
true_positions = pd.read_csv(folder+"/true_positions.csv",index_col=0)["0"].values
A_true = pd.read_csv(folder+"/true_adjacency.csv",index_col=0).values
n_steps=len(df.index)
```

```python
lambda_true,c_true,r_true  = ground_truth_params
```

```python
positions_history = df.iloc[:,(2+len(ground_truth_params)):]
```

```python
start_positions = positions_history.iloc[0]
```

```python
fig,ax = plt.subplots()
ax.scatter(true_positions, start_positions)
ax.plot([0,2*np.pi],[0,2*np.pi])
ax.set_xlabel("true positions")
ax.set_xlabel("sampled positions at start time")
```

```python
fig,ax = plt.subplots()
ax.scatter(true_positions, positions_history.iloc[-1])
ax.plot([0,2*np.pi],[0,2*np.pi])
ax.set_xlabel("true positions")
ax.set_xlabel("sampled positions at endtime")
```

```python
pos = np.vstack([true_positions, positions_history.iloc[-1]]).T
G = nx.from_numpy_array(A_true)

nx.draw_networkx_edges(G,pos=pos,alpha=0.1)
plt.scatter(pos[:,0], pos[:,1])
```

```python
fig,ax = plt.subplots()
ax.plot(df["2"])
ax.plot([0,n_steps],[lambda_true,lambda_true])
ax.set_xlabel("time")
ax.set_ylabel("lambda")
```

```python
fig,ax = plt.subplots()
ax.plot(df["3"])
ax.plot([0,n_steps],[c_true,c_true])
ax.set_xlabel("time")
ax.set_ylabel("c")
```

```python
fig,ax = plt.subplots()
ax.plot(df["4"])
ax.plot([0,n_steps],[r_true,r_true])
ax.set_xlabel("time")
ax.set_ylabel("r")
```
